```python
import numpy as np 
import pandas as pd

#for visualization
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('Mall_Customers.csv',encoding='unicode_escape')
```

```python
df.head()
```

```python
sns.heatmap(df.isna())
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.countplot(x='Genre', data=df);
plt.title('Gender Distribution');
```

```python
df.hist('Age', bins=35);
plt.title('Age Distribution');
plt.xlabel('Age');
```

```python
plt.hist('Age', data=df[df['Genre'] == 'Male'], alpha=0.5, label='Male');
plt.hist('Age', data=df[df['Genre'] == 'Female'], alpha=0.5, label='Female');
plt.title('Age by Gender Distribution');
plt.xlabel('Age');
plt.legend();
```

```python
df.hist('Annual Income (k$)');
plt.title('Annual Income Distribution in Thousands of Dollars');
plt.xlabel('Thousands of Dollars');
```

```python
plt.hist('Annual Income (k$)', data=df[df['Genre'] == 'Male'], alpha=0.5, label='Male');
plt.hist('Annual Income (k$)', data=df[df['Genre'] == 'Female'], alpha=0.5, label='Female');
plt.title('Income by Gender Distribution');
plt.xlabel('Income (Thousands of Dollars)');
plt.legend();
```

```python
male_customers = df[df['Genre'] == 'Male']
female_customers = df[df['Genre'] == 'Female']

# Print the average spending score for men and women
print(male_customers['Spending Score (1-100)'].mean())
print(female_customers['Spending Score (1-100)'].mean())
```

```python
sns.scatterplot('Age', 'Annual Income (k$)', hue='Genre', data=df);
plt.title('Age to Income, Colored by Gender');
```

```python
sns.heatmap(df.corr(), annot=True)
```

```python
sns.scatterplot('Age', 'Spending Score (1-100)', hue='Genre', data=df);
plt.title('Age to Spending Score, Colored by Gender');
```

```python
sns.heatmap(female_customers.corr(), annot=True);
plt.title('Correlation Heatmap - Female');
```

```python
sns.heatmap(male_customers.corr(), annot=True);
plt.title('Correlation Heatmap - Male');
```

```python
sns.lmplot('Age', 'Spending Score (1-100)', data=female_customers);
plt.title('Age to Spending Score, Female Only');
```

```python
sns.scatterplot('Annual Income (k$)', 'Spending Score (1-100)', hue='Genre', data=df);
plt.title('Annual Income to Spending Score, Colored by Gender');
```
